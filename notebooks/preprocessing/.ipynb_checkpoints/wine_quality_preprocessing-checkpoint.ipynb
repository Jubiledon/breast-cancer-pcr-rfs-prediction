```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ucimlrepo import fetch_ucirepo 
```

```python
# fetch dataset 
wine_quality = fetch_ucirepo(id=186) 
  
# data (as pandas dataframes) 
X = wine_quality.data.features 
y = wine_quality.data.targets 
  
# metadata 
#print(wine_quality.metadata) 
  
# variable information 
#print(wine_quality.variables)

#check the columns
print(f"Feature columns \n {X.columns}")
print(f"\nTarget column:\n {y.columns}") 
```

```python
df = pd.concat([X, y], axis=1)

df.info()
```

```python
df.head()
```

```python
print("Mssing val: ", df.isnull().sum())
```

```python
before = len(df)
df = df.drop_duplicates().reset_index(drop=True)
after = len(df)
print(f"Dropped {before - after} duplicate rows.")
```

```python
X = df.drop(columns=['quality'])
y = df['quality']

print(X.shape, y.shape)
```

### Feature Normalisation


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Xs = scaler.fit_transform(X)
fig,ax=plt.subplots(1,figsize=(20,8))
sns.boxplot(data=Xs,ax=ax) 

```

### PCA


```python
from sklearn.decomposition import PCA
feature_names = list(X.columns)
pca = PCA(n_components=10)
Xs_pca = pca.fit_transform(Xs)
PCA_df = pd.DataFrame()
PCA_df['PCA_1'] = Xs_pca[:,0]
PCA_df['PCA_2'] = Xs_pca[:,1]
PCA_df['quality'] = y.values
```

```python
Xs_pca
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='PCA_1', y='PCA_2', hue='quality', palette='viridis', data=PCA_df)
plt.title('PCA projection colored by Wine Quality')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

```python
import joblib
import pandas as pd

#save splits to CSV
#create csv of what you need. i.e X, normalised X (Xs), Dimentional reduced X (Xs_pca)

pd.DataFrame(Xs, columns=X.columns).to_csv("X_scaled.csv", index=False)
pd.DataFrame(X, columns=X.columns).to_csv("X.csv", index=False)
pd.DataFrame(Xs_pca).to_csv("X_pca.csv", index=False)
y.to_csv("y.csv", index=False)

#save scaler
joblib.dump(scaler, "scaler.pkl")
```
